```python
import requests

# test using SACOG bldg coords
lat = 38.576861105263355
long = -121.48826664994633
coords_sacog = [[long, lat]] # ORS takes in coords as (long, lat)

coords_default = [[8.6757369817168, 49.400736250782884]] # central Heidelberg, default map area for ORS docker installation

api_key = input("enter file path of the ORS API text file: ")
mode = "cycling-regular"  # "driving-car", "foot-walking", "cycling-regular"
isoctype = "distance" # "time", "distance" 
travel_range = 0.5 # enter time in minutes, distance in miles
batch_sz = 2 # how many points you want to make isos around for each request you send to ORS--max is 5.


# url = f'https://api.openrouteservice.org/v2/isochrones/{mode}'
url = f'http://localhost:8080/ors/v2/isochrones/{mode}'

coords = coords_sacog

#=======RUN SCRIPT=============
 # [[lat, long]]
    
range_mts = travel_range * 1609.34

body = {"locations":coords, "range":[range_mts], "range_type":isoctype}

headers = {
'Accept': 'application/json, application/geo+json, application/gpx+xml, img/png; charset=utf-8',
'Authorization': api_key,
'Content-Type': 'application/json; charset=utf-8'
}

call = requests.post(url, json=body, headers=headers)

call.json()
```
