```python
# Load tabular data from ESRI-hosted API
import geopandas as gpd
import plotly.express as px

# get url for api
api_url = r"https://services1.arcgis.com/5NARefyPVtAeuJPU/arcgis/rest/services/Crime/FeatureServer/0/query?where=1%3D1&outFields=*&outSR=4326&f=json"

# response = requests.request("POST", api_url)

# r_json = response.json()

# api url returns GeoJSON, which can be read directly into geopandas geodataframe
gdf = gpd.read_file(api_url)

gdf['lon'] = gdf['geometry'].x.values
gdf['lat'] = gdf['geometry'].y.values

fig = px.density_mapbox(gdf, lat='lat', lon='lon',
                        center=dict(lat=38.6, lon=-121), zoom=10,
                        radius=7,
                        mapbox_style="open-street-map")

fig.show()
```
